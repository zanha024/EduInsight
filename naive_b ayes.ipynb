```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("newData2.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = df.drop(columns='mental_activities')
df = df.drop(columns='time-management')
df.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for column in df.columns:
    # Check if the column contains non-numeric data
    if df[column].dtype == 'object':
        # Initialize label encoder for the column
        label_encoders[column] = LabelEncoder()
        # Encode the values in the column and replace them with integers
        df[column] = label_encoders[column].fit_transform(df[column])
df.head(20)
```

```python
X=df.drop(columns='learning_strategy')
y=df['learning_strategy']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
X_train_std = ss.fit_transform(X_train_imputed)
X_test_std = ss.transform(X_test_imputed)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_model = GaussianNB()
nb_model.fit(X_train_std, y_train)
y_pred = nb_model.predict(X_test_std)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
